in case of parking spaces we can assume that if not given means not present so we will fill it with 0

```python
df['numParkingSpaces']=df['numParkingSpaces'].fillna(0)
```

yearBuilt we can fill it with the mod of all the years as it should be around it

```python
df['yearBuilt']=df['yearBuilt'].fillna(df['yearBuilt'].mode())
```

lotside and living area can be considered as mean as it is the best option rather than excluding the rows

```python
df['lotSize']=df['lotSize'].fillna(df['lotSize'].mean())
df['livingArea']=df['livingArea'].fillna(df['livingArea'].mean())
```
